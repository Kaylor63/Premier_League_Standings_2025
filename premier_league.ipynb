```python
import json

# If standings_list is a JSON string, parse it first
if isinstance(standings_list, str):
    standings_list = json.loads(standings_list)

rows = []
column_names = [
    "season", "position", "team", "points",
    "all_played", "all_win", "all_draw", "all_lose",
    "all_goals_for", "all_goals_against", "goal_diff"
]

for club in standings_list:
    season = 2025
    rank = club['stats']['rank']
    team = club['team']['name']
    points = club['stats']['points']
    all_played = club['stats']['gamesPlayed']
    all_win = club['stats']['wins']
    all_draw = club['stats']['ties']
    all_lose = club['stats']['losses']
    all_goals_for = club['stats']['goalsFor']
    all_goals_against = club['stats']['goalsAgainst']
    goal_diff = club['stats']['goalDifference']

    tuple_of_club_records = (season, rank, team, points, all_played, all_win, all_draw, all_lose, all_goals_for, all_goals_against, goal_diff)
    rows.append(tuple_of_club_records)
```

```python
# Set MYSQL_USER if not set
if not MYSQL_USER:
    MYSQL_USER = "pl_app_writer"  # <-- Change this to your actual MySQL username if different

db_connection = connector.connect(
    host=MYSQL_HOST,
    port=MYSQL_PORT,
    user=MYSQL_USER,
    password=MYSQL_PASSWORD,
    database=MYSQL_DATABASE,
    connection_timeout=15,
    autocommit=False,
    raise_on_warnings=True
)

cur = db_connection.cursor()
print(f"[SUCCESS] Connection to MySQL DB '{MYSQL_DATABASE}' was successful.")
```

```python
sql_table = "standings"
cur.execute(f"SHOW TABLES LIKE '{sql_table}'")

if cur.fetchone() is None:
    raise SystemExit(f"[ERROR] Table '{sql_table}' does not exist in DB '{MYSQL_DATABASE}'")
else:
    print(f"[SUCCESS] Table '{sql_table}' exists in DB '{MYSQL_DATABASE}'")
```

```python
#Start the UPESERT operation

table_cols = [ "season", "position", "team", "points",
    "all_played", "all_win", "all_draw", "all_lose",
    "all_goals_for", "all_goals_against", "goal_diff"]

standings_df = df[table_cols]
```

```python
standings_records_tuples = standings_df.itertuples(index=False, name=None)

list_ofstandings_records_tuples = list(standings_records_tuples)    
```

```python
UPSERT_SQL = f"""
INSERT INTO {sql_table} (season, position, team, points,
    all_played, all_win, all_draw, all_lose, all_goals_for, all_goals_against, goal_diff)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s) as src
ON DUPLICATE KEY UPDATE
    position     = src.position,
    points  = src.points,
    all_played  = src.all_played,
    all_win     = src.all_win,
    all_draw    = src.all_draw,
    all_lose    = src.all_lose,
    all_goals_for = src.all_goals_for,
    all_goals_against = src.all_goals_against,
    goal_diff   = src.goal_diff;
"""
```

```python
no_of_records = len(list_ofstandings_records_tuples)
```

```python
try:
    cur.executemany(UPSERT_SQL, list_ofstandings_records_tuples)
    db_connection.commit()
    print(f"[SUCCESS] UPSERT operation completed successfully. {no_of_records} records processed.")
except connector.Error as e:
    db_connection.rollback()
    print(f"[ERROR] UPSERT operation failed: {e}")
finally:
    cur.close()
    db_connection.close()
    print(f"All database connections closed.")
```
